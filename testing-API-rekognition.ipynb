## Label detection

### primera prueba

```python
data_video = client.start_label_detection(Video={"S3Object": {"Bucket": "fiep-ctv","Name": "PRUEBAS/outdoors.mp4"}},MinConfidence=70)
display(data_video)
```

```python
response_video = client.get_label_detection(JobId=data_video['JobId'])
response_video
```

### Prueba tiempo
Para video de 1 minuto el job succeded en 1:30, se obtuvo respuesta en 1:40.

```python
data_video = client.start_label_detection(Video={"S3Object": {"Bucket": "fiep-ctv","Name": "PRUEBAS/reina_flow_1min.mp4"}},MinConfidence=70)
display(data_video)
```

```python
response_video = client.get_label_detection(JobId=data_video['JobId'])
response_video
```

## Label moderation

```python
label_mod = client.start_content_moderation(Video={"S3Object": {"Bucket": "fiep-ctv","Name": "PRUEBAS/outdoors.mp4"}},MinConfidence=70)
display(label_mod)
```

```python
response_label_mod = client.get_content_moderation(JobId=label_mod['JobId'])
response_label_mod
```

### Prueba tiempo

Para un video de 1 minuto se obtuvieron resultados en 43 segundos.

```python
label_mod = client.start_content_moderation(Video={"S3Object": {"Bucket": "fiep-ctv","Name": "PRUEBAS/reina_flow_1min.mp4"}},MinConfidence=70)
display(label_mod)
```

```python
response_label_mod = client.get_content_moderation(JobId=label_mod['JobId'])
response_label_mod
```

## Segment detection

```python
shot_det = client.start_segment_detection(Video={"S3Object": {"Bucket": "fiep-ctv","Name": "PRUEBAS/outdoors.mp4"}},
                                         SegmentTypes=['SHOT', 'TECHNICAL_CUE'])
display(shot_det)
```

```python
response_shot_det = client.get_segment_detection(JobId=shot_det['JobId'])
response_shot_det
```

### Prueba tiempo
Para un video de 1 minuto se obtuvieron resultados en 36 segundos.
